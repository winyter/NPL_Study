## Linear Algbra(线性代数)

```python
points = [
    (0, 0), 
    (0.5, 0), 
    (0.5, 6.44),
    (6, 0),
    (6, 8),
    (5.5, 8),
    (5.5, 1.58),
    (0, 8)
]
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([0, 1], [1, 2])
```

```python
def draw_segment(segments):
    for i,seg in enumerate(segments):
        _next = segments[(i + 1) % len(segments)]
        x1, y1 = seg
        x2, y2 = _next
        plt.plot([x1, x2], [y1, y2])
```

```python
draw_segment(points)
```

```python
import numpy as np
```

```python
points_matrix = np.array(points)
```

```python
transform = np.array([
    [1, 0.4],
    [0.8, 1]
])
```

```python
transformed = np.dot(points_matrix, transform)
draw_segment(transformed)
```

## Probability

```python
import random
```

```python
def dice():
    return random.randint(1, 6)
```

```python
dice()
```

```python
occurences = [
    (i, j) for i in range(1, 7) for j in range(1,7)
]
```

```python
occurences
```

## 至少出现一次3的概率是多少

```python
event = [(a, b) for a,b in occurences if a==3 or b==3]
```

```python
def probability(event, occurences):
    return len(event) / len(occurences)
```

```python
probability(event, occurences)
```

## 如果第一次出现的是1， 那么第二次出现3的概率是多少？

```python
first_is_one = [
    (a, b) for a,b in occurences if a==1
]
```

```python
first_is_one
```

```python
second_is_three = [
    (a, b) for a,b in first_is_one if b==3
]
```

```python
second_is_three
```

```python
probability(second_is_three, first_is_one)
```

## Centeral Limitation Theroy 中心极限定律

一个随机事件（比如骰子），如果重复的足够多时，它会呈现正态分布

在深度学习中，通常用来初始化模型函数的权值

```python
%matplotlib inline 
import matplotlib.pyplot as plt
from statistics import mean
from random import normalvariate, triangular, choice, vonmisesvariate, uniform

def normal(mu=0, sigma=1): return random.normalvariate(mu, sigma)

def repeated_hist(dist, n=10**6, bins=100):
    "Sample the distribution n times and make a histogram of the results."
    samples = [dist() for _ in range(n)]
    plt.hist(samples, bins=bins, normed=True)
    plt.title('{} (μ = {:.1f})'.format(dist.__name__, mean(samples)))
    plt.grid(axis='x')
    plt.yticks([], '')
    plt.show()
```

```python
repeated_hist(normal, n=1000)
```

```python
repeated_hist(normal, n=10000)
```

```python
repeated_hist(normal, n=50000)
```

```python
def sum10dice(): return sum(random.randint(1, 6) for _ in range(10))
```

```python
sum10dice()
```

```python
repeated_hist(sum10dice, 10)
```

```python
repeated_hist(sum10dice, 100)
```

```python
repeated_hist(sum10dice, 1000)
```

```python
repeated_hist(sum10dice, 10000)
```

```python
repeated_hist(sum10dice, 1000000)
```
