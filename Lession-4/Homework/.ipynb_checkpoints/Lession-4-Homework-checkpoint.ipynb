```python
def all_path(dirname):

    result = []#所有的文件

    for maindir, subdir, file_name_list in os.walk(dirname):

        #print("1:",maindir) #当前主目录
        #print("2:",subdir) #当前主目录下的所有目录
        #print("3:",file_name_list)  #当前主目录下的所有文件

        for filename in file_name_list:
            apath = os.path.join(maindir, filename)#合并成一个完整路径
            result.append(apath)

    return result

print(all_path("/Users/winyter/repository/zhwiki/AA"))
```

```python
# 获取语料
def extract(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as fd:
        lines = fd.readlines()
        for i, line in enumerate(lines):
            #print(line)
            lists = line.strip('\n').split('\t')
            #print(lists)
            sentence = "".join(lists)
            sentence = re.sub(r'<[^>]+>', '', sentence)
            sentence = re.sub(r'[0-9]', '', sentence)
            sentence = re.sub(r'[a-z]', '', sentence)
            sentence = re.sub(r'[A-Z]', '', sentence)
            #print(sentence)
            with open(output_file, 'a+', encoding='utf-8') as wp:
                wp.write(sentence)
                wp.close()
        fd.close()
```

```python
# 加载停用词，使用了哈工大停用词
def stopWords(path):
    stopwords = []
    with open(path, 'r', encoding="UTF-8") as f:
        lines = f.readlines()
        for line in lines:
            stopwords.append(line.strip())
    return stopwords
```

```python
stopWords('/Users/winyter/repository/stopwordslist.txt')
```

```python
# 分词
def cutwords(corpus, stopwords_path, cut_save_path):
    stopwords = stopWords(stopwords_path)
    cutCorpus = open(cut_save_path, 'a', encoding='utf-8')
    with open(corpus, 'r', encoding='utf-8') as f:
        text = f.readlines()
        for i, sentence in enumerate(text):
            #if i%10 == 0:
                #print("cutted {0}".format(i))
            sentence = list(jieba.cut(sentence))
            sentence_cut = []
            for word in sentence:
                if word not in stopwords:
                    sentence_cut.append(word)
            cutCorpus.write(" ".join(sentence_cut))
        del text
        f.close
        cutCorpus.close()
```

```python
cutwords('/Users/winyter/repository/NPL-Study/Lession-4/resource/extracttest.txt','/Users/winyter/repository/stopwordslist.txt','/Users/winyter/repository/NPL-Study/Lession-4/resource/cut_result.txt')
```

```python
def run(data_path, corpus_path, stopwords_path, cutCorpus, output1, output2):
    # 提取语料
    extract(data_path, corpus_path)
    print("---extarct completed---")
    # 分词
    cutwords(corpus_path, stopwords_path, cutCorpus)
    print("---cutwords completed---")
    # 训练词向量
    sentences = word2vec.Text8Corpus(cutCorpus)
    model = word2vec.Word2Vec(sentences, size=300, window=10, min_count=5, iter=20)
    model.save(output1)
    model.wv.save_word2vec_format(output2, binary=False)
    
    print(model.most_similar('你好'))
    print()
    print(model.most_similar('股票'))
    print()
    print(model.most_similar('腾讯'))
    print()
    
    return model
```

```python
def tsne_plot(model1):
    "Creates and TSNE model and plots it"
    labels = []
    tokens = []

    for word in model1.wv.vocab:
        tokens.append(model1[word])
        labels.append(word)

    tsne_model = TSNE(perplexity=40, n_components=2, init='pca', n_iter=100, random_state=23)
    new_values = tsne_model.fit_transform(tokens)

    x = []
    y = []
    for value in new_values:
        x.append(value[0])
        y.append(value[1])

    plt.figure(figsize=(16, 16)) 
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
                     xytext=(5, 2),
                     textcoords='offset points',
                     ha='right',
                     va='bottom')
    plt.show()
```

```python
if __name__ == '__main__':
    data_path = '/Users/winyter/repository/zhwiki/AA/wiki_30'
    corpus_path = '/Users/winyter/repository/NPL-Study/Lession-4/resource/extracttest.txt'
    stopwords_path = '/Users/winyter/repository/stopwordslist.txt'
    cutCorpus = '/Users/winyter/repository/NPL-Study/Lession-4/resource/cut_result.txt'
    output1 = '/Users/winyter/repository/NPL-Study/Lession-4/resource/vec.model'
    output2 = '/Users/winyter/repository/NPL-Study/Lession-4/resource/word2vec_format'
    model = run(data_path, corpus_path, stopwords_path, cutCorpus, output1, output2)
    
```
