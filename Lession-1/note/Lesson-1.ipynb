```python
adj_grammar = """
Adj* => null | Adj Adj*
Adj => 蓝色的 | 好看的 | 小小的
"""
```

```python
grammar = {}
```

```python
for line in adj_grammar.split('\n'):
    if not line.strip(): continue
    
    exp, stmt = line.split('=>')
    grammar[exp.strip()] = [s.split() for s in stmt.split('|')]  # s. 后面不使用 strip 的原因是把右边部分的空格也当成分隔符进行切割，下面在使用时，会把这个子列表内的值做&运算
    
# 以上程序的思路是建立一个数据结构，把adj_grammar转换为这个数据结构，"=>"左边的为可扩展的东西(即字典 key)，右边为可扩展内容(即字典 value)
```

```python
grammar
```

```python
def generate(gram, target):
    if target in gram:   # target could be expanded，即 target 在 gram 的 key 中
        new_expanded = random.choice(gram[target])    #随机获取任意一个此 key 的 value
        return ''.join(generate(gram, t) for t in new_expanded)   #递归调用，直到返回的是一个不可扩展的值
    else:
        return target
# 举例说明，如果 t 返回的是“蓝色的”，那么，t 会被作为 target 递归传入 generate ,而“蓝色的“不是 generate 的 key，所以，generate 就会直接打印出这个值
```

```python
generate(grammar, "Adj*")
```

```python
def create_grammar(grammar_str):
    grammar = {}
    for line in grammar_str.split('\n'):
        if not line.strip(): continue
        exp, stmt = line.split('=>')
        grammar[exp.strip()] = [s.split() for s in stmt.split('|')]
    return grammar
```

```python
example_grammar = create_grammar(simple_grammar)
example_grammar
```

```python
def generate(gram, target):
    if target in gram:
        expanded = random.choice(gram[target])
        new_expanded = [generate(gram, t) for t in expanded]
        return ' '.join(ne for ne in new_expanded if ne != 'null')
    else:
        return target
```

### Data Driven （数据驱动）
- 我们的目标是，希望能做一个程序，然后，当输入的数据变化的时候，我们的程序不需要重写。Generalization
- AI？如何能自动化解决问题，我们找到一个方法之后，输入变了，我们的方法不用变

```python
programming = """
stmt => if_exp | while | assignment
assignment => var = var
if_exp => if (var) {stmt}
while => while (var) {var}
var => chars number
chars => char | char char
char => a | b | c | d | e
number => 1 | 2 | 3"""
```

以上所有的内容，就是 Rule based 的应用，即我使用规定的语法，来生成一个句子，但是，我们如何判断一个句子是合理还是不合理呢？
Rule based 仅能支持通过一定的语法生产特定的句子
如何去判断句子的合理性，就需要用到：Language Model，这个模型是一个函数，输入为一个句子，输出为(0,1)范围的数字，这个数字越大，句子越合理

而这个函数的计算，则需要基于概率（简单来说，就是我们通过程序，统计一个单词出现在一个句子某个位置上的概率，概率越高意味着我们人类用的越多，也就意味着这个句子越合理），这就引出了 AI Paradigm 的第二部分：Probability Based

### 2、Probability Based
- 一句话为什么我们说不对 => 不合常理 => 也就是小概率

### Language Model
$$ language\_model(string) = Probability(String)  \in (0,1) $$
$$ Pro(w_1 w_2 w_3 w_4) = Pr(w_1 | w_2 w_3 w_4) * Pr(w_2 | w_3 w_4) * Pr(w_3 | w_4) * Pr(w_4) $$
$$ Pro(w_1 w_2 w_3 w_4) \sim Pr(w_1 | w_2 ) * Pr(w_2 | w_3 ) * Pr(w_3 | w_4) * Pr(w_4) $$

- 第一个公式的含义：语言模型，就是输入一个 string，然后看这个 string 的概率，这个概率在 0和 1 之间(不会等于 0 或 1)
- 第二个公式就是计算一句话（含有 w1,w2,w3,w4 这四个单词）出现的概率
- 第三个公式实际上是第二个公式的近似简化，我们设想一下，如果这个公式有 100 个单词，那么使用第二个公式计算概率虽然很准确，但实际上很复杂，甚至很难实现，为了保证效率，我们可以引入第三个公式，做一个近似值的计算，而第三个公式实际上已经很简单了。整个两个公式可以理解为：第二个公式中，我们假设的是 w1 在 w2w3w4 出现的条件下出现的概率，也就是说，w1 出现和 w2w3w4 都相关，但是，我们可以假设，w1 出现的概率仅和靠近自己的 w2 有关，w2 仅和靠近自己的 w3 有关，以此类推

- 扩展：概率论中 | 符号的含义：P(A|B)，这是条件概率中的内容，这表示在 B 条件下 A 的概率，即 $ Count(A)/Count(B) $
    - 由此，上面的 $ Pr(w_1 | w_2 w_3 w_4) = Count(w_1)/Count(w_2 w_3 w_4) $ 需要注意，这里的 count(w1) 是指 w1 和 w2w3w4 一起出现的次数，也就是在 w2w3w4 出现的条件下出现 w1 的概率

```python
TOKEN = [str(t) for t in TOKEN]
```

```python
TOKEN_2_GRAM = [''.join(TOKEN[i:i+2]) for i in range(0,len(TOKEN)-2)]
```

```python
TOKEN_2_GRAM
```

```python
words_count_2 = Counter(TOKEN_2_GRAM)
```

```python
def prob_2(word1,word2):
    if word1 + word2 in words_count_2: return words_count_2[word1+ word2] / len(TOKEN_2_GRAM)
    else:
        return 1 / len(TOKEN_2_GRAM)
```

```python
prob_2("我们","在")
```

```python
def get_probablity(sentence):
    words = cut(sentence)
    
    sentence_pro = 1
    for i, word in enumerate(words[0:-1]):
        next_ = words[i+1]
        probablity = prob_2(word, next_)
        sentence_pro *= probablity
    return sentence_pro
```

```python
get_probablity("你好世界")
```
